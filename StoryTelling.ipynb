```python
# Applying loop and slicing to get three months sum then add the previous sum to the next three months sum
# saving the values to 'deaths' list
strt = 0
end = +3
deaths = []

for i in range (4):
    cum_deaths = df['New_deaths'][strt:end].sum()    
    deaths.append (cum_deaths)
    end += 3
```

```python
# calling each value of a quarter and dividing it to the total value to get percentage
dts = [round (dt / max(deaths), 2) for dt in deaths]
dts
```

```python
# we have axhline for plotting the load bar
# we have normal load bar starting from '0' and ending on '9'
# here we have load bar starting from 0.5 and ending on 0.8 means we have 0.3 as total
# so we are multiplying out total deaths ratio to this proportion
prop = [(0.5 + dt * 0.3) for dt in dts]
prop
```

```python
# named the plot as ax1 ax2, and so on, 
# we have taken 04 rows and one column with 6 and 8 dimension
fig, [ax1, ax2, ax3, ax4] = plt.subplots(4, 1, figsize = (6, 8))
axes = [ax1, ax2, ax3, ax4]

# Plotting the graph to all columns through loop
# we have month on x axis and deaths on y axis
# Set params = 0 on left and bottom sides
# remove the spines for all sides
for ax in axes:
    ax.plot(df['Month'], df['New_deaths'], color = '#b30000', alpha = 0.4)
    ax.set_xticklabels([])
    ax.set_yticklabels([])
    ax.tick_params(left = 0, bottom = 0)
    ax.spines[['left', 'right', 'top', 'bottom']].set_visible (False)


# Data divided in to 04 quarters and plotting quarter wise
# line width enhanced to 2.5
ax1.plot(df['Month'][:3], df['New_deaths'][:3], color = '#b30000', linewidth = 2.5)
ax2.plot(df['Month'][2:6], df['New_deaths'][2:6], color = '#b30000', linewidth = 2.5)
ax3.plot(df['Month'][5:9], df['New_deaths'][5:9], color = '#b30000', linewidth = 2.5)
ax4.plot(df['Month'][8:], df['New_deaths'][8:], color = '#b30000', linewidth = 2.5)

# Editing the text to the graph
ax1.text(0.5,-60,0) # Horizontal, vertical then text
ax1.text(11.5,2380,2247, alpha = 0.6)
ax1.text(1.2,-350,'Jan-Mar', color = '#b30000', weight = 'bold', rotation = 4)
ax1.text(2.4, 220,'77 Deaths', color = '#b30000', weight = 'bold', rotation = 70)
ax1.text(0.5, 3500, 'The virus have killed {:,} people'.format (df.New_deaths.sum()), \
         color = '#330000', size = 12, weight = 'bold')

ax2.text(0.5,-60,0)
ax2.text(11.5,2380,2247, alpha = 0.6)
ax2.text(3.5, 400,'Apr-Jun', color = '#b30000', weight = 'bold', rotation = 70)
ax2.text(4.1, 1000,'4077 Deaths', color = '#b30000', weight = 'bold', rotation = -36)

ax3.text(0.5,-60,0)
ax3.text(11.5,2380,2247, alpha = 0.6)
ax3.text(7, 400,'Jul-Sep', color = '#b30000', weight = 'bold', rotation = 10)
ax3.text(4.5, 900,'2528 Deaths', color = '#b30000', weight = 'bold', rotation = -36)

ax4.text(0.5,-60,0)
ax4.text(11.2, 1000,'Oct-Dec', color = '#b30000', weight = 'bold', rotation = 60)
ax4.text(9.7, 1000,'4223 Deaths', color = '#b30000', weight = 'bold', rotation = 50)

# in below loop axes as plot names
# prop showing the total deaths multiplying to 0.3 as we have bar starting from 5 and ending to 8 so we have 0.3 s total
# deaths are showing the total deaths by quarter
# dts showing the percent of deaths by quarter to total deaths
for ax, propor, death, percent  in zip (axes, prop, deaths, dts):
    ax.axhline(y = 2100, xmin = 0.5, xmax = 0.8, linewidth = 6, color = '#b30000', alpha = 0.3)
    ax.axhline(y = 2100, xmin = 0.5, xmax = propor, linewidth = 6, color = '#b30000')
    ax.text(8, 1750, format(death, ','), color = '#b30000', weight = 'bold')
    ax.text(8.2, 2300, str(int(percent * 100)) + '%', color = '#b30000', weight = 'bold')
    
plt.show()
```
